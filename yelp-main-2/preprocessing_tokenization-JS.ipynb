## 1. Preprocessing

```python
## Obtaining the reviews 
text = test_sample['text']
```

```python
## Created a DataFrame with the reviews
text_df = pd.DataFrame(text)
text_df
```

```python
import re

## Preprocessing the text data to maintain english language characters
def contains_only_english(text):
    return bool(re.match(r'^[a-zA-Z\s]*$', text))

## Removing special characters and space characters
text_df = pd.DataFrame(text_df.text.str.lower())
text_df = pd.DataFrame(text_df.text.str.replace('\n\n',''))
text_df = pd.DataFrame(text_df.text.str.replace(r'/[^a-zA-Z0-9]', ""))
text_df = pd.DataFrame(text_df['text'].str.replace('.', ''))
text_df = pd.DataFrame(text_df['text'].str.replace(',', ''))
text_df = pd.DataFrame(text_df['text'].str.replace('!', ''))

# remove non english
text_df = text_df[text_df['text'].apply(contains_only_english)]
```

First, we used the function (^^[a-zA-Z\s]*$). This regex checks for the reviews that contain—only English words. During the analysis of the notebook, we came across some reviews that were in different languages. However, we decided to keep only those reviews that were in the English language. As part of the preprocessing process, we converted all the text to lowercase and removed the unique and space characters to ensure the text data was clean for the following models and steps. Then, we removed the lines of text that were not alphanumeric or spaces, and finally, we filtered them by the English characters only.

## 3. Bag of Words

```python
pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## Removing stop words with Count Vectorizer
cv = CountVectorizer(stop_words='english')
cv.fit(text_df.text)

print("Number of 'tokens'", len(cv.vocabulary_))
cv.vocabulary_
```

```python
## Sampling the data to only 50,000 attributes due to Memory issues
shrunk = text_df.head(50000)

## Obtaining the Bag of Words DataFrames with the Feature Names
dtm = cv.transform(shrunk['text'])
bow = pd.DataFrame(dtm.toarray(), columns = cv.get_feature_names_out())
```

```python
bow 
```

We first removed the stop words to initialize the Bag of Words method. Then, we fit the reviews into the Count Vectorizer to build the vocabulary and prepare the transformation of the text into the vector representation of Bag of Words. Finally, we decided to sample the dataset due to computing and memory issues. We eventually created a document term called "dtm" to sparse the matric, in which rows represent documents and columns the unique words. 

## 4. Generate n-grams

```python
# Getting a smaller subset of the data

smaller_sample= text_df.sample(n=1000, random_state=42)
smaller_sample
```

```python
cv_ngrams=CountVectorizer(stop_words='english', ngram_range=(2,2))
cv_ngrams.fit(smaller_sample['text'])
dtm_ngrams=cv_ngrams.transform(smaller_sample['text'])
```

```python
# Converting to a data frame
bow_ngrams=pd.DataFrame(dtm_ngrams.toarray(),columns=cv_ngrams.get_feature_names_out())
```

```python
bow_ngrams
```

Used a smaller sample of the data to make these next steps less computationally expensive. Generated a bag of 2-grams from the sample to be able to capture enough context from the text without being too sparse. This way, it is also easier to get more insight from the sample than from just one word reviews.

## 5. Computing Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix= cosine_similarity(bow_ngrams)
similarity_matrix
```

Finally, computed the cosine similarity matrix on the smaller sample we took of our data to explore similarities within our reviews data sample. We can see that we have a diagonal of all 1's, which means that each review is perfectly similar to itself. The zeros in the rest of the matrix suggest that every review is unique and doesn't match up with the rest.

## 6. Recommendation System

```python
# Getting a smaller subset of the data

smaller_sample= test_sample.sample(n=1000, random_state=42)
smaller_sample
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(smaller_sample['name'])

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_recommendations(index, cosine_sim=cosine_sim):
    sim_scores = list(enumerate(cosine_sim[index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11] 

    restaurant_indices = [score[0] for score in sim_scores]
    
    return smaller_sample.iloc[restaurant_indices]

recommendations = get_recommendations(0)
recommendations = pd.DataFrame(recommendations)
```

```python
recommendations = recommendations.drop(columns=['Unnamed: 0', 'business_id', 'user_id', 'text', 'date', 'funny', 'review_id', 'yelping_since', 'submitted_reviews', 'address', 'latitude', 'longitude','average_user_stars','tip_review','attributes','cool'])
```
