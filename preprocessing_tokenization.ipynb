## 1. Preprocessing

```python
## Obtaining the reviews 
text = test_sample['text']
```

```python
## Created a DataFrame with the reviews
text_df = pd.DataFrame(text)
text_df
```

```python
import re

## Preprocessing the text data to maintain english language characters
def contains_only_english(text):
    return bool(re.match(r'^[a-zA-Z\s]*$', text))

## Removing special characters and space characters
text_df = pd.DataFrame(text_df.text.str.lower())
text_df = pd.DataFrame(text_df.text.str.replace('\n\n',''))
text_df = pd.DataFrame(text_df.text.str.replace(r'/[^a-zA-Z0-9]', ""))
text_df = pd.DataFrame(text_df['text'].str.replace('.', ''))
text_df = pd.DataFrame(text_df['text'].str.replace(',', ''))
text_df = pd.DataFrame(text_df['text'].str.replace('!', ''))

# remove non english
text_df = text_df[text_df['text'].apply(contains_only_english)]
```

First, we used the function (^^[a-zA-Z\s]*$). This regex checks for the reviews that contain—only English words. During the analysis of the notebook, we came across some reviews that were in different languages. However, we decided to keep only those reviews that were in the English language. As part of the preprocessing process, we converted all the text to lowercase and removed the unique and space characters to ensure the text data was clean for the following models and steps. Then, we removed the lines of text that were not alphanumeric or spaces, and finally, we filtered them by the English characters only.

## 3. Bag of Words

```python
pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## Removing stop words with Count Vectorizer
cv = CountVectorizer(stop_words='english')
cv.fit(text_df.text)

print("Number of 'tokens'", len(cv.vocabulary_))
cv.vocabulary_
```

```python
## Sampling the data to only 50,000 attributes due to Memory issues
shrunk = text_df.head(50000)

## Obtaining the Bag of Words DataFrames with the Feature Names
dtm = cv.transform(shrunk['text'])
bow = pd.DataFrame(dtm.toarray(), columns = cv.get_feature_names_out())
```

We first removed the stop words to initialize the Bag of Words method. Then, we fit the reviews into the Count Vectorizer to build the vocabulary and prepare the transformation of the text into the vector representation of Bag of Words. Finally, we decided to sample the dataset due to computing and memory issues. We eventually created a document term called "dtm" to sparse the matric, in which rows represent documents and columns the unique words.
